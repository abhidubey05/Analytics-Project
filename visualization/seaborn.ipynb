```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = sns.load_dataset('iris')
df
```

```python
df.isnull().sum()
```

```python
sns.scatterplot(df, x ='sepal_length', y='sepal_width', hue='species')
plt.title('sepal_length by petal_length')
plt.show()
```

```python
sns.histplot(df, x='petal_length', bins=30,kde=True, color='red')
plt.title('petal_length')
plt.show()
```

```python
sns.boxplot(df, y='sepal_width', x='species')
plt.title('sepal_width by boxplot')
plt.show()
```

```python
sns.violinplot(df, x='petal_length', y='species',split=True, palette='Set2')
plt.title('petal_length by species')
plt.show()
```

```python
sns.countplot(df, x='species', palette='viridis')
plt.title('species of countplot')
plt.show()
```

```python
sns.relplot(df, x='sepal_length', y='sepal_width', palette='spring')
plt.title('sepal_length by sepal_width')
plt.show()
```

```python
sns.swarmplot(df, x='sepal_width', y='sepal_length', palette='viridis')
plt.title('sepal_width by sepal_length')
plt.show()
```
